```python
data.describe().T
```

### Расчет жизненного цикла  платформ

```python
years=data.pivot_table(index=['platform',], values='year_of_release', aggfunc=['min','max'])
years.plot(kind='bar', title='Старт и конец продаж',figsize=(8,5), grid=True, legend=False)
plt.ylim(2003,2017)
#years.columns
years.columns=years.columns.droplevel(1)
```

```python
years_live=[]
for i in years['max'],years['min']:
        years_live.append(years[min]-years[max])
        
years_live=np.average(years_live).round()
years_live
```

Средний жизненный цикл платформ составляет 6 лет, хотя встречались и неудачные платформы, жизненный цикл которых составил всего 1-3 года (DC, GBA, GC). Платформы с наиболее длинным жизненным циклом в 10-12 лет - это PC, PS3,Wii,X360. 

Для дальнейшего анализа и прогноза продаж на 2017 год, будем брать только платформы, у которых дата релиза игр попадает в диапазон, равный половине среднего жизненного цикла. 

```python
sales_platforms=data[data['year_of_release']>=(2016-years_live/2)].sort_values(by='all_sales', ascending=False).reset_index(drop=True)
```

### Продажи по платформам за актуальный период

```python
sales_platforms.pivot_table(index='year_of_release',columns='platform', 
                            values='all_sales', aggfunc='sum').plot.bar(rot=0,grid=True,
                                                                        figsize=(13,5));
```

```python
top=sales_platforms.groupby(['platform'])['all_sales'].sum().sort_values(ascending=False).reset_index().head()
top.plot(x='platform',kind='bar',figsize=(8,5));
```

```python
topp=top['platform']
top_sales_platforms=sales_platforms.query('platform in @topp').reset_index(drop=True)
top_sales_platforms
```

```python
top_sales_platforms['genre'].unique()
```

### ТОП платформ по глобальным продажам за актуальный период



На данном графике видим платформы, лидирующие по продажам в данном периоде. Можем отследить в какой стадии жизненного цикла находится каждая из них. Например, PS3 и X360 критично близки к концу своего жизненного цикла, потому как вышли обновленные платформы PS4 и XOne. Так как в нашем датасете представленны не полные данные за 2016 год, - мы не можем с уверенностью сказать, что продажи этих платформ падают, потому как пик графика приходится как раз на 2015 год. Скорее всего эти две платформы (PS4 и XOne) будут иметь топовые либо очень высокие продажи в 2017 году. Продажи по платформе 3DS также падают, однако не так стремительно, по графику видно, что падение в год составляет примерно 10 млн копий, и на конец 2016 года, предположительно, продажи будут на уровне 20 млн копий, следовательно, в 2017 году продажи, хоть и снизятся, но все еще будут на уровне 10-20 млн копий.

### Влияние оценок пользователей и критиков на продажи платформы PS3  


```python
plt.figure(figsize=(16, 6)) 
mask = np.triu(np.ones_like(p_ps3.corr(), dtype=np.number)) 
heatmap = sns.heatmap(p_ps3.corr(), mask=mask, vmin=-1, vmax=1, annot=True, cmap='BrBG') 
heatmap.set_title('Корреляция');
```

```python
p_ps3.plot(x='all_sales', y='critic_score', kind='scatter',figsize=(18,8), grid=True,  color='r')
plt.title("Зависимость продаж игр PS3 от оценок критиков")
plt.show();
```

```python
p_ps3.plot(x='all_sales', y='user_score', kind='scatter', grid=True, figsize=(18,8), color='g')
plt.title("Зависимость продаж игр PS3 от оценок пользователей")
plt.show();
```


- Проверив коэффициент корреляции Пирсона и построив диаграмму рассеивания для определения зависимости роста продаж определенных игр от оценки критиков, следует отметить, что зависимости прямой нет, но это не значит, что пользователи не смотрят на оценку критиков, возможно, дело в том, что очень много игр данной платформы имеют высокие оценки.

- Судя по диаграмме рассеивания и коэффициенту корреляции Пирсона,- нет прямой зависимости уровня продаж игр платформы PS3 от оценки пользователей. Однако, можно заметить, что подавляющее большенство игр данной платформы имеют оценку пользователей выше среднего.



### Влияние оценок пользователей и критиков на продажи разных платформ

```python
platforms = ['3DS','PS4','XOne','PC']
new_platforms = sales_platforms.query('platform in @platforms')
for row in ['critic_score','user_score']:
    sns.relplot(data = new_platforms, x=row, y='all_sales', kind='scatter', col='platform', color='m')
    print('Платформа:','Корреляция all_sales и',row, new_platforms.groupby('platform')['all_sales'].corr(new_platforms[row]))
```

По данным коэффициента корреляции и диаграммам рассеивания можно отметить, что ситуация с влиянием оценок пользователей и критиков на продажи всех платформ сопоставима с нашими выводами по платформе PS3: нет прямой зависимости продаж от оценок.

## Проверка гипотез

### Гипотеза о платформах  Xbox One и PC

Нулевой гипотезой примем равенство: средние пользовательские рейтинги платформ Xbox One и PC одинаковые. 

Альтернативной гипотезой будет утверждение, что они не равны. 

```python
rate_users_pc=sales_platforms.query('platform == "PC"').dropna(subset=['user_score']).reset_index(drop=True)
rate_users_xone=sales_platforms.query('platform == "XOne"').dropna(subset=['user_score']).reset_index(drop=True)
```

```python
rate_users_pc['user_score'].median()
```

```python
rate_users_xone['user_score'].median()
```

```python
alpha=.01
results = st.ttest_ind(rate_users_xone['user_score'], rate_users_pc['user_score'], equal_var=False)
print('p-значение:', results.pvalue)
```

```python
if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

После проведения Т-теста для двух выборок по платформе XOne и PC, получили значение p-value и сравнили его с заданным уровнем альфа. Так как доверительный интервал выбран 99% - значение альфа установили 0.01. Результаты показали, что мы не можем отвергнуть нулевую гипотезу о том, что средние пользовательские рейтинги платформ XOne и PC равны. Значит, рейтинги равны.

### Гипотеза о жанрах Sport  и Action

Сформулируем нулевую гипотезу: Средние пользовательские рейтинги жанров Sport и Action равны.

Тогда альтернативной гипотезой будет, что они разные.


```python
user_rate_action=data.query('genre == "Action"').dropna(subset=['user_score']).reset_index(drop=True)
user_rate_sports=data.query('genre == "Sports"').dropna(subset=['user_score']).reset_index(drop=True)
```

```python
user_rate_action['user_score'].median()
```

```python
user_rate_sports['user_score'].median()
```

```python
alpha=.01
results = st.ttest_ind(user_rate_action['user_score'], user_rate_sports['user_score'], equal_var=False)
print('p-значение:', results.pvalue)
```

```python
if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

Проведя Т-тест мы смогли отвергнуть нулевую гипотезу о равенстве пользовательских ретингов жанров  Sport и Action. Таким образом мы подтвердили нашу гипотезу из задания: средние пользовательские рейтинги жанров Sport и Action разные. Так как выборки у нас независимые и разных размеров, дисперсию оцениваем по отдельности для каждой выборки. Уровень альфа также установили в 1%.
